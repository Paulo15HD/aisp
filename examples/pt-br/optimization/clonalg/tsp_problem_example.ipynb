# Otimização com o Algoritmo de Seleção Clonal (CLONALG)

## 1. Introdução

### Problema do Caixeiro Viajante (TSP)

Neste notebook, aplica o ClonAlg ao **Problema do Caixeiro Viajante (Traveling Salesman Problem - TSP)**.

O problema consiste em encontrar o caminho mais curto que visita cada cidade exatamente uma vez e retorna à cidade de origem.

### Importação das bibliotecas necessárias

```python
# Importando o Algoritmo de Seleção Clonal (CLONALG)
from aisp.csa import Clonalg

# Bibliotecas para manipulação de dados e cálculos numéricos
import numpy as np
import random
from itertools import combinations
from scipy.spatial import distance

# Visualização de dados
import matplotlib.pyplot as plt
import networkx as nx
```

## 3. Definição do problema

Cria 15 cidades com coordenadas aleatórias e define a função ``tsp_fitness``.

```python
# Configuração do problema
n_cities = 15
cities = []
random.seed(123)

# Gerar coordenadas aleatórias para as cidades
for i in range(n_cities):
    x = random.uniform(0, 100)
    y = random.uniform(0, 100)
    cities.append((x, y))

cities = np.array(cities)

# Função de fitness para TSP
def tsp_fitness(solution):
    total_distance = 0
    n = len(solution)

    for i in range(n):
        current_city = cities[solution[i]]
        next_city = cities[solution[(i + 1) % len(solution)]]
        total_distance += distance.euclidean(current_city, next_city)
    return total_distance

print(f"Número de cidades: {n_cities}")
print("Coordenadas das cidades:")
for i, city in enumerate(cities):
    print(f"Cidade {i}: ({city[0]:.2f}, {city[1]:.2f})")
```

## 4. Configuração do algoritmo

```python
clonalg = Clonalg(
    problem_size=n_cities,
    N=100,
    selection_size=10,
    rate_clonal=50,
    rate_hypermutation=0.05,
    n_diversity_injection=15 ,
    feature_type="permutation-features",
    seed=121
)

clonalg.register('affinity_function', tsp_fitness)
```

## 5. Execução da otimização

```python
clonalg.optimize(500, 20)

if clonalg.best_cost is not None:
    print('Melhor distância encontrada:', clonalg.best_cost)
    print('Melhor rota:', clonalg.best_solution)
```

## 6. Resultado

```python
print(clonalg.get_report())
```

### 6.1 Evolução do melhor ao longo das gerações

```python
plt.figure(figsize=(10, 6))
plt.plot(clonalg.cost_history, marker='o', linewidth=2, markersize=4)
plt.title('Evolução da Distância ao Longo das Interações', fontsize=16)
plt.xlabel('Interações', fontsize=12)
plt.ylabel('Distância Total', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
def plot_all_routes_with_best(cities, best_route, title):
    G = nx.Graph()
    for i, (x, y) in enumerate(cities):
        G.add_node(i, pos=(x, y))
    all_edges = list(combinations(range(len(cities)), 2))
    G.add_edges_from(all_edges)
    
    # Adicionar arestas da melhor rota para destaque
    best_route_extended = list(best_route) + [best_route[0]]  # Fechar o ciclo
    best_edges = [(best_route_extended[i], best_route_extended[i+1]) for i in range(len(best_route_extended)-1)]
    
    pos = nx.get_node_attributes(G, 'pos')
    
    plt.figure(figsize=(12, 8))
    
    # Desenhar as arestas
    nx.draw_networkx_edges(G, pos, edgelist=all_edges, edge_color='lightgray', alpha=0.5)
    nx.draw_networkx_edges(G, pos, edgelist=best_edges, edge_color='blue', width=2)
    
    nx.draw_networkx_nodes(G, pos, node_size=300, node_color='red')
    
    nx.draw_networkx_labels(G, pos, font_size=12, font_color='black')
    
    plt.title(title, fontsize=16)
    plt.xlabel('X', fontsize=12)
    plt.ylabel('Y', fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.axis('equal')
    plt.show()

if clonalg.best_solution is not None:
    plot_all_routes_with_best(cities, clonalg.best_solution, f'Melhor Rota Encontrada - (Distância Total: {clonalg.best_cost:.2f})')
```
